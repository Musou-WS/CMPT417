```python
import csv
import matplotlib.pyplot as plt
import numpy as np
```

```python
dg_2_approx = "2-approx.csv"
dg_brute_force = "dg.csv"
wdg_linear = "wdg_linear.csv"
wdg_round = "wdg_round.csv"
iwdg = "iwdg.csv"
lazy = "lazy.csv"
```

```python
dg_2_approx_time = []
dg_2_approx_generate = []
dg_2_approx_expand = []
with open(dg_2_approx, 'r', encoding="utf-8") as csvfile:
    dg_2_approx_reader = csv.reader(csvfile)
    for i,row in enumerate(dg_2_approx_reader):
        if i != 0 and i != 42:
            dg_2_approx_time.append(float(row[2]))
            dg_2_approx_generate.append(int(row[3]))
            dg_2_approx_expand.append(int(row[4]))
```

```python
# plt.plot(x, dg_brute_force_time, linestyle='', marker='.')
plt.plot(x, dg_brute_force_time, label='Brute-Force')
plt.plot(x, dg_2_approx_time, label='2-Approximation')
plt.title('time')
plt.legend()
```
