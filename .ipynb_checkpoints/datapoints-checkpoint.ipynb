```python
import csv
import matplotlib.pyplot as plt
import numpy as np
```

```python
dg_2_approx = "2-approx.csv"
dg_brute_force = "dg.csv"
wdg_linear = "wdg_linear.csv"
wdg_round = "wdg_round.csv"
iwdg = "iwdg.csv"
lazy = "lazy.csv"
```

```python
dg_2_approx_time = []
dg_2_approx_generate = []
dg_2_approx_expand = []
with open(dg_2_approx, 'r', encoding="utf-8") as csvfile:
    dg_2_approx_reader = csv.reader(csvfile)
    for i,row in enumerate(dg_2_approx_reader):
        if i != 0:
            dg_2_approx_time.append(row[2])
            dg_2_approx_generate.append(row[3])
            dg_2_approx_expand.append(row[4])
```

```python
dg_brute_force_time = []
dg_brute_force_generate = []
dg_brute_force_expand = []
with open(dg_brute_force, 'r', encoding="utf-8") as csvfile:
    dg_brute_force_reader = csv.reader(csvfile)
    for i,row in enumerate(dg_brute_force_reader):
        if i != 0:
            dg_brute_force_time.append(row[2])
            dg_brute_force_generate.append(row[3])
            dg_brute_force_expand.append(row[4])
```

```python
x = np.linspace(1, 50, 50)
```

```python
# plt.plot(x, dg_brute_force_time, linestyle='', marker='.')
plt.legend()
plt.plot(x, dg_brute_force_time, label='Brute-Force')
plt.plot(x, dg_2_approx_time, label='2-Approximation')
```
